```python
import numpy as np
from scipy.interpolate import BSpline
```

```python
#Generating sample data
x= np.linspace(0,10,10)
y= np.sin(x)
```

```python
#Defining degree of the BSpline and the number of knots
degree=1
num_knots = len(x)+degree+1
```

```python
#Generate knot vector
knots = np.linspace(x[0],x[-1],num_knots)
```

```python
#Perform BSpline interpolation
bspline = BSpline(knots, y, degree)
```

```python
#Evaluate the BSpline at desired positions
x_interpolated= np.linspace(x[0],x[-1],100)
y_interpolated = bspline(x_interpolated)
```

```python
#Plot the original data and the interpolated curve
import matplotlib.pyplot as plt
plt.scatter(x,y,label="Originial Data")
plt.plot(x_interpolated, y_interpolated, label="Interpolated curve")
plt.legend()
plt.show()
```
